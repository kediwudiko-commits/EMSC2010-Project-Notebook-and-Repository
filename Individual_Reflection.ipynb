Refelcting on this project, I initially selected this topic because south-eastern Australia's exposure to extreme rainfall events felt both scientifically interesting and practically given the region's history of devastating floods.

My first unexpected challenge came early in the preprocessing stage, where a data handling error caused the fitted Normal distribution to return physically impossible parameter values (μ about 1000 mm). Debugging this taught me the importance of always validating outputs against real-world expectations before proceeding.

A more significant methodological challenge acros when I initially attempted to implement a full Generalised Extreme Value (GEV) model with a free shape parameter (xi). While theoretically more flexible, this produced over 1800 MCMC divergences, indicating severe sampling instability. After reflection, I recognised that constraining the shape parameter to zero, effectively adopting the Gumbel distribution, it was not only more numerically stable but also more appropriate given the limited record lengths available. This decision highlighted a key lesson: model complexity should be justified by the data, not imposed for its own sake.

My working process also became disorganised as I iterated through these different approaches. Ultimatley, I made the decision to rebuild my notebook from scratch in a clean, well-structured file. While this cost time, it significantly improved the reproducibility and clarity of my final analysis.

The transition from Maximum Likelihood Estimation to full Bayesian framework was particularly enlightening. Unlike MLE, which provides a single point estimate, Bayesian inference allowed me to propagate parameter uncertainty directly into risk assessments, producing credible intervals rather than false precision. The wide HDI for Sydney's return period estimates ([3.2, 28.8] years for a 150 mm event) honestly reflects the genuine uncertainty inherent in extreme value analysis with short records.

The key strength of this analysis is its explicit quantification of uncertainty. Its main shortcoming is the stationarity assumption, the model treats historical patterns as fixed, ignoring potential shifts driven by climate change or ENSO variability. If I were to extend this work, incorporating time varying covariates would be the natural next step.
